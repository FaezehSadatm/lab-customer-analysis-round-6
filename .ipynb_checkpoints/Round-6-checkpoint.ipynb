```python
def log_transform(x):
    if np.isfinite(x) and x!=0:
        return np.log(x)
    else:
        return np.NAN
```

```python
df['TOTAL_CLAIM_AMOUNT'] = df['TOTAL_CLAIM_AMOUNT'].apply(log_transform).fillna(df['TOTAL_CLAIM_AMOUNT'].mean())
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
y = df['TOTAL_CLAIM_AMOUNT']
X = df.drop(['TOTAL_CLAIM_AMOUNT'], axis=1)
```

```python
X_num = X.select_dtypes(include = np.number)
X_cat = X.select_dtypes(include = 'object')
```

```python
encoder = OneHotEncoder(handle_unknown='error', drop='first')
encoder.fit(X_cat)
```

```python
encoded = encoder.transform(X_cat).toarray()
X = np.concatenate([X_num, encoded], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = LinearRegression() 
model.fit(X_train,y_train)
```

```python
predictions  = model.predict(X_test)
```

```python
r2_score(y_test, predictions), mean_squared_error(y_test, predictions, squared=False), mean_absolute_error(y_test, predictions)
```
